```python
X_train,X_teste,y_train,y_teste = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
score = clf.score(X_teste, y_teste)
```

```python
predict = clf.predict(X_teste)
```

```python
y_teste = y_teste.reset_index(drop=True)
predict = pd.DataFrame(predict)
df_results = pd.concat([predict, y_teste], axis=1)
df_results.columns = ['predict','Real']
display(df_results)
```
